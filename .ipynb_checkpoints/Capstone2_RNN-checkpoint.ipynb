```python
stu_df = pd.read_csv('./2018_stu_encode_no_normal.csv', sep=',', header=0, keep_default_na=False, na_values=[''])
stu_df = stu_df.reset_index(drop=True)
# feature_names = stu_df.columns.to_list()
# X = stu_df.loc[:, ('admissions_score', 'school_type_encod', 'major_correlation_encod', 'GPA',
#                    'admissions_approach_encode', 'province_encod', 'title_encoding', 'type_encoding', 'major_encod',
#                    'gender_encod')]
# normalization_list = ['admissions_score', 'school_type_encod', 'major_correlation_encod', 'GPA',
#                       'admissions_approach_encode', 'province_encod', 'title_encoding', 'type_encoding',
#                       'major_encod', 'gender_encod']
X = stu_df.loc[:, ('admissions_score', 'school_type_encod', 'major_correlation_encod', 'GPA', 'admissions_approach_encode', 'province_encod')]
normalization_list = ['admissions_score', 'school_type_encod', 'major_correlation_encod', 'GPA', 'admissions_approach_encode', 'province_encod']
# 数据Min-Max归一化
for i in normalization_list:
    Max = np.max(X[i])
    Min = np.min(X[i])
    X[i] = (X[i] - Min) / (Max - Min)
X = X.round(2)
X = X.iloc[:, :].values
# X = stu_df.iloc[:, :10].values
Y = stu_df.iloc[:, 10].values
Y1 = np.array([1 if i >= 0.6 else 0 for i in Y])
```

```python
#CNN
#Construct the model
from tensorflow.python.keras.losses import categorical_crossentropy
from keras.layers import Conv2D,Conv1D
n_neurons=7
start_time=time.time()
model = Sequential()
model.add(Conv1D(n_neurons,kernel_size=1))
# model.add(Dense( activation="tanh", kernel_initializer="uniform", units=64, activity_regularizer=regularizers.l2(0.01)))
# model.add(Dense( activation="tanh", kernel_initializer="uniform", units=32, activity_regularizer=regularizers.l2(0.01)))
model.add(Dense( activation="softmax", kernel_initializer="uniform", units=3))
model.compile(loss=categorical_crossentropy, optimizer='adam',metrics=['accuracy'])
```

```python
#LSTM
#Construct the model
from tensorflow.python.keras.losses import categorical_crossentropy
n_neurons=7
start_time=time.time()
model = Sequential()
model.add(LSTM(n_neurons, input_shape=(X_train.shape[1],X_train.shape[2])))
model.add(Dense( activation="tanh", kernel_initializer="uniform", units=64, activity_regularizer=regularizers.l2(0.01)))
model.add(Dense( activation="tanh", kernel_initializer="uniform", units=32, activity_regularizer=regularizers.l2(0.01)))
model.add(Dense( activation="softmax", kernel_initializer="uniform", units=3))
model.compile(loss=categorical_crossentropy, optimizer='adam',metrics=['accuracy'])
```

```python
#CNN+LSTM
#Construct the model
from tensorflow.python.keras.losses import categorical_crossentropy
from keras.layers import Conv2D,Conv1D,MaxPooling1D,Activation
n_neurons=7
start_time=time.time()
model = Sequential()
model.add(Conv1D(filters=64,kernel_size=1,activation='relu'))
model.add(MaxPooling1D(1))
model.add(Activation('relu'))
model.add(keras.layers.LSTM(units=32,activation='tanh'))
model.add(Dense( activation="softmax", kernel_initializer="uniform", units=3))
model.compile(loss=categorical_crossentropy, optimizer='adam',metrics=['accuracy'])
```
